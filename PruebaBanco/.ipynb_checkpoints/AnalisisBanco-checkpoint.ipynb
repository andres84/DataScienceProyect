# **Analisis prueba banco**

#### Requerimientos:

1.   Consolidar en un DataFrame los archivos Cartera_31_DIC_2020 y Cartera_31_DIC_2020 guardados
2.   Utilizando el DataFrame consolidado completar los datos que se encuentran en el archivo P&F

#### Nota

Tener en cuenta las siguientes formulas:
* ***ICV(Indicador de cartera vencida) =  Saldo vencido / saldo total***
* ***VARIACION = (situacion final - situacion inicial) / situacion inicial***

```python
#Despues de realizar la normalizacion y el analisis de los archivos Cartera_31_DIC_2020 y Cartera_31_DIC_2020 se procede a realizar la carga de los datos
#cargar y consolidar en un dataframe los archivos

#importar librerias
import pandas as pd
from glob import glob as gg

#creacion de una lista vacia
consolidado = []
#creacion de un iterador para buscar los archivos a consolidar
for archive in gg('Car*.xlsx'):# a traves de gg cargo los archive
    dfcon = pd.read_excel(archive)# creo el dataframe
    consolidado.append(dfcon)#agrego el dataframe a la lista
dfcon = pd.concat(consolidado)#uno la data de los achive
```

```python
#visualizacion de los datos unificados
dfcon.head()
```

```python
#revisar el tipo de datos y si alguna celda esta vacia
dfcon.info()
```

```python
#Resolviendo el requerimiento 2
#calcular saldo_total, saldo_vencido, Obligaciones_Total y Obligaciones_vencidas
#agrupamos los datos por regional y por mo_fecha_de_proceso
totales = dfcon.groupby(['regional', 'mo_fecha_de_proceso']).sum()
totales
```

```python
#con el dataframe de los primeros totales creamos la columna ICV para realizar el carculo de este
totales['ICV'] = totales['Saldo_vencido'] / totales['Saldo_total']
totales
```

```python
# Siguiendo con la resolucion del requerimiendo 2, ahora se procede a clcular la variacion
# teniendo en cuenta la formula de variacion procedemos a resolver la primera parte de la formula con la funcion .diff()
# calcular diff de las columna Saldo_total, Saldo_vencido y ICV creando una columna y guardando los resultado en cada una
totales['diff_Saldo_total'] = totales.Saldo_total.diff()
totales['diff_Saldo_vencido'] = totales.Saldo_vencido.diff()
totales['diff_ICV'] = totales.ICV.diff()
totales
```

```python
#despues de haber clculado la primera parte de la formula de variacion procedemos a resolverla en su totalidad
#resolvemos la variacion de cada columna segun el requerimmiento

#variacion Saldo_total
varSaldo_total = totales.diff_Saldo_total/totales.Saldo_total.shift(1)*100
totales['Variacion_Saldo_total'] = varSaldo_total/100 #creacion de la columna

#variacion Saldo_total
varSaldo_vencido = totales.diff_Saldo_vencido/totales.Saldo_vencido.shift(1)*100
totales['Variacion_Saldo_vencido'] = varSaldo_vencido/100 #creacion de la columna

#variacion ICV
varICV = totales.diff_ICV/totales.ICV.shift(1)*100
totales['Variacion_ICV'] = varICV/100 #creacion de la columna

#visualizar el dataframe
totales
```
