### formulas

- ICV(Indicador de cartera vencida) =  Saldo vencido / saldo total
- variacion = (Datofinal - Datoinicial) / Datoinicial

```python
#carga, union de datos y creacion de dataframe

import pandas as pd
from glob import glob as gg

contenido = []

for archive in gg('car*.xlsx'):
    dfconsolidado = pd.read_excel(archive)
    contenido.append(dfconsolidado)
dfconsolidado = pd.concat(contenido)
```

```python
#cálculo totales
totales = dfconsolidado.groupby(['regional', 'mo_fecha_de_proceso']).sum()
totales
```

```python
# cálcular ICV
totales['ICV'] = totales.Saldo_vencido / totales.Saldo_total
totales
```

```python
# calcular diff de saldo_total, saldo_vencido y ICV
totales['diff_saldo_total'] = totales.Saldo_total.diff()
totales['diff_saldo_vencido'] = totales.Saldo_vencido.diff()
totales['diff_ICV'] = totales.ICV.diff()
totales
```

```python
#cálculo variacion saldo_total, saldo_vencido y ICV
varSaldoTotal = totales.diff_saldo_total / totales.Saldo_total.shift(1)*100
totales['variacion_saldo_total'] = varSaldoTotal / 100

varSaldoVencido = totales.diff_saldo_vencido / totales.Saldo_vencido.shift(1)*100
totales['variacion_saldo_vencido'] = varSaldoVencido / 100

varICV = totales.diff_ICV / totales.ICV.shift(1)*100
totales['variacion_ICV'] = varICV / 100

totales
```

```python
totales.to_excel('dfTotales.xlsx')
```
