# Unir varios archivos txt

```python

import pandas as pd
from glob import glob as gg
import numpy as numpy

consolidarCartera = []
for f in gg('*.csv'):
    df = pd.read_csv(f, sep = ';')
    consolidarCartera.append(df)
       
df = pd.concat(consolidarCartera, ignore_index =True)
df.head(10)
```

```python
consolidarCart = []
for f in gg('*.txt'):
    df = pd.read_csv(f,encoding ='latin1', sep = ';')
    consolidarCart.append(df)
       
df = pd.concat(consolidarCartera, ignore_index =True)
df.head(10)
```
