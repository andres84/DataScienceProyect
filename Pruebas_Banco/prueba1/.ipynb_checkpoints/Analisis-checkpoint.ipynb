## Antes de proceder a la carga de los archivos se debe realizar la revision y normalizacion de estos

### NOTA

Antes de crear el codigo para unir los archivos con pandas y glob se debe realizar un analisis
de los archivo txt y pasarlos a csv manualmente, tener en cuenta las columnas tipo numericas
se debe convertir a celdas calculables reemplazando el punto por la coma

```python
import pandas as pd
from glob import glob as gg #ayuda a la carga de multiples archivos

consolidadoCartera = []
for i in gg('car*.xlsx'): #se crea un for para que busque y recorra los archivos
    dfconso = pd.read_excel(i)#lee cada archivo recorrido
    consolidadoCartera.append(dfconso)#agrega a la lista los archivos encontrados
dfconso = pd.concat(consolidadoCartera)#concatena los archivos encontrados
```

## Nota en caso que aplique

En la consulta anterior se observa que las columnas Saldo_total y Saldo_vencido son de tipo objeto

Se deben transformar a tipos de datos numericos para porder realizar cálculos matematcos

```python
dfconso['Saldo_total'] = dfconso['Saldo_total'].astype('int64')
dfconso['Saldo_vencido'] = dfconso['Saldo_vencido'].astype('int64')
dfconso['Obligaciones_vencidas'] = dfconso['Obligaciones_vencidas'].astype('int64')
```
