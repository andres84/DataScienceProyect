```python
#prueba de dataframe para calcular una variacion
import pandas as pd

matriz = {'año':['2019', '2020', '2021', '2022'],
          'Censo_2020':[1000, 2000, 500, 300]}
datos = pd.DataFrame(matriz)
datos
```

```python
datos['dif'] = datos.Censo_2020.diff()
datos
```

```python
dife = datos.dif/datos.Censo_2020.shift(1)*100
datos['Var'] = dife/100
```
