#### 1. Crear un dataframe donde se cruce la informacion de personas con la infrmacion de edades para obtener el nombre y la edad de cada persona

```python
import pandas as pd
from glob import glob as gg
#cargar las hojas individualmente

dfEdades = pd.read_excel("DATOS.xlsx", sheet_name='EDADES')
dfPersonas = pd.read_excel('DATOS.xlsx', sheet_name='PERSONAS')

dfPersonas.head()
```

```python
dfEdades.head(15)
```

```python
# realizar cruce entre los dataframe dfpersonas y dfEdades para extraer la edad de cada persona
dfEdadesPersonas = pd.merge(dfPersonas, dfEdades[['CEDULA', 'EDAD']], how='left')
dfEdadesPersonas.head(10)
```

```python
# 2. Consolidar las tablas de ventas
dfconsolidado = pd.read_excel('DATOS.xlsx', sheet_name = ['VENTAS 2020', 'VENTAS 2021'])
dfconVentas = pd.concat(dfconsolidado)
dfconVentas.head(50)
```

```python
# 3. Generar un data frame que agrupe el total de ventas por mes
totalesMes = dfconVentas.groupby(['MES', 'AÑO']).sum()
totalesMes
```

```python
# 4. Generar un data frame que agrupe el total de ventas por año
totalesAño = dfconVentas.groupby('AÑO').sum()
totalesAño
```

```python
# 5. Generar una columna donde se concatene el mes con el año separado por /

dfcontieneMES_AÑO = pd.DataFrame(dfconVentas)
dfcontieneMES_AÑO['Union_MES_AÑO'] = dfcontieneMES_AÑO.MES.map(str) + '/' + dfcontieneMES_AÑO.AÑO.map(str)
#dfcontieneMES_AÑO.info()
dfcontieneMES_AÑO.head()
```

```python
# 6. Crear dataframe que almacene el total de ventas
dfTotalVentas = pd.DataFrame(dfconVentas)
dfTotalVentas['TotalVentas'] = dfTotalVentas.VENTAS.sum()
dfTotalVentas
#dfTotalVentas.iloc['TotalVentas']
```
